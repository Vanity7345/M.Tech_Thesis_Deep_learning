```python
# prompt: now use the protac_cnv dataset and keep data for dc 50 and dmax which have atleat one value either for dmax and dc50. also keep data which have both values for dc50 and dmax

from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Specify the file path
file_path = "/content/drive/My Drive/mtech_project/protac_cnv.csv"

# Load the CSV file
df = pd.read_csv(file_path)

# Check for correct column names (case-sensitive and spaces)
print(df.columns)

# Filter data
# Keep rows with at least one value for 'DC50 (nM)' or 'Dmax (%)'
filtered_df = df.dropna(subset=['DC50 (nM)', 'Dmax (%)'], how='all')

# Save the filtered data
output_path = "/content/drive/My Drive/mtech_project/filtered_protac_cnv.csv"
filtered_df.to_csv(output_path, index=False)

print(f"Filtered data saved to {output_path}")
print(filtered_df.head())
```
